```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

x2 = np.arange(0, 4.5, 0.5)

plt.figure(figsize = (5,3), dpi = 120)

plt.plot(x, y, label='2x', color = "maroon", markersize = 8, marker = ".", linewidth = 2, linestyle = "--")
plt.plot(x2, x2**2)

plt.title("Our first graph", fontdict = {"fontname": "Comic Sans MS", "fontsize" : 20})
plt.xlabel("X label")
plt.ylabel("Y label")

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.legend()

plt.savefig('my_graph.png')

plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1,4,6]

bars = plt.bar(labels, values)
bars[0].set_hatch('/')
bars[1].set_hatch('*')
bars[2].set_hatch('o')

plt.show()
```
